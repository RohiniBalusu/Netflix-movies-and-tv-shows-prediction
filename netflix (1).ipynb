# Netflix Movies and TV Shows Prediction by Scikit-Learn Library

### Objective:
The objective of applying machine learning algorithms using the Scikit-Learn library on a dataset of Netflix movies and TV shows prediction is to  develop machine learning models that can predict the success and popularity of movies and TV shows on the platform. By analyzing features like genre, director, release year, and type of content, the goal is to forecast viewer engagement, preferences, and trends. This enables Netflix to optimize its recommendation system, improve content curation, and make data-driven decisions for future content development and marketing strategies, ultimately enhancing user experience and retention.


### About Dataset:
Within this dataset, you will find a comprehensive collection of data Netflix Movies and TV Shows Prediction Dataset typically this dataset allows for the application of machine learning algorithms based on historical data. By analyzing patterns, trends, and correlations between these features, content development includes data about the content available on the platform, offering to  help  understand various attributes of each movies or TV shows. Which includes show_id,type,title,director,cast,country,data_added,release_year,rating,duration,listed_in.

### Dataset Glossary (Column-Wise):
- **Show_id**:A unique identifier for each show or movie. The format could be "s1", "s2", "s3", etc.
- **Type**: Specifies whether the content is a TV Show or a Movie.
- **Title**:The name of the movie or TV show. It identifies the content within the dataset and helps in categorizing or searching for a specific title.
- **Director**:This column contains the names of the directors who created the movie or TV show.
- **Cast**:The primary actors or performers in the content.The list of actors or performers who star in the movie or TV show
- **Country**: The country or countries where the content originated or was produced. This helps identify the geographical origin of the content.   
- **Data_added**: This column shows the exact date when the content was made available on Netflix.The format typically follows the YYYY-MM-DD structure                     (Year-Month-Day), which makes it easy to track specific dates.
- **Release_year**:The year when the movie or TV show was originally released. This helps in analyzing content trends over time and understanding the a                     content available on Netflix.
- **Rating**: The content’s rating, which might include MPAA ratings (like G, PG, R, etc.) for movies or ratings for TV shows.
- **Duration**: - The total runtime of the movie (in minutes) or the number of episodes for a TV show.
- **Listed_in**: This column serves to provide additional classification for each movie or TV show based on its genre, type, or content theme. (such as                    Horror Movies, International Movies, Dramas, etc.)

```python
import plotly.express as px
def generate_rating_df(df):
    rating_df = df.groupby(['rating', 'target_ages']).agg({'type': 'count'}).reset_index()
    rating_df = rating_df[rating_df['type'] != 0]
    rating_df.columns = ['rating', 'target_ages', 'counts']
    rating_df = rating_df.sort_values('target_ages')
    return rating_df


rating_df = generate_rating_df(df)
fig = px.bar(rating_df, x='rating', y='counts', color='target_ages', title='Ratings of Movies And TV Shows Based On Target Age Groups',  labels={'counts':'COUNT', 'rating':'RATINGS', 'target_ages':'TARGET AGE GROUPS' })
fig.show()
#Ratings of Movies And TV Shows Based On Target Age Groups
```

```python
def correlation(dataset, threshold):
    col_corr = set() # Set of all the names of deleted columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if  (corr_matrix.iloc[i, j] >= threshold) :
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
    return col_corr
```
